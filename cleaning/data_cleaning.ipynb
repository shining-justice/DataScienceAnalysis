# Merging

```python
df = charts_df.merge(store_info_df, on='app_id', how='inner')
```

```python
df.head()
```

```python
df.info()
```

# Cleaning

## App ID, rank

```python
df.drop(columns=['app_id', 'rank'], inplace=True)
```

# Name

```python
df[df.name.str.contains('Beta|Test|Demo|Server', case=False, na=False)]
```

```python
df = df[~df.name.str.contains('Beta|Test|Demo|Server', case=False, na=False)].reset_index(drop=True)
```

## Current players, 24-hour peak, all time peak

```python
columns = ['current_players', '24h_peak', 'all_time_peak']

df[columns] = (df[columns].apply(lambda s: s.str.replace(',', '', regex=False).astype(int)))
```

## App type

```python
df.app_type.unique()
```

```python
df = (df.loc[df['app_type'] == 'Game'].reset_index(drop=True))
```

```python
df.drop(columns=['app_type'], inplace=True)
```

## Supported systems

```python
df.drop(columns=['supported_systems'], inplace=True)
```

## Release date

```python
df['release_date_clean'] = df.release_date.str.replace('–', '', regex=False).str.strip()
df['release_date_dt'] = pd.to_datetime(df.release_date_clean, format='%d %B %Y %H:%M:%S UTC', errors='coerce')
df.release_date = df.release_date_dt.dt.strftime('%d.%m.%Y')

df.drop(columns=['release_date_clean', 'release_date_dt'], inplace=True)
```

## Hardware categories

```python
df.drop(columns=['hardware_categories'], inplace=True)
```

# Accessibility categories  

```python
df.drop(columns=['accessibility_categories'], inplace=True)
```

# Price

```python
df.drop(columns=['price_usd_current', 'price_eur_current', 'price_eur_lowest', 'price_cis_current', 'price_cis_lowest'], inplace=True)
```

```python
df.price_usd_lowest = (df.price_usd_lowest.str.extract(r'\$(\d+(?:\.\d+)?)').astype(float))
```

```python
df[df.tags.str.contains('Free to Play', case=False, na=False)]
```

```python
df.loc[df.tags.str.contains('Free to Play', case=False, na=False), 'price_usd_lowest'] = 0
```

```python
df = df.rename(columns={'price_usd_lowest': 'price'})

df = df[['name', 'developer', 'publisher', 'release_date', 'tags', 'categories', 
         'rating_percent', 'reviews_count', 'price', 'current_players', 
         '24h_peak', 'all_time_peak']]
```

```python
df.to_csv('data_cleaned.csv', index=False)
```
